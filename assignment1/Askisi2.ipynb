```python
import random
import numpy as np
import scipy as sp
import pandas as pd
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

df  = pd.read_csv('data.csv')
df.plot(x='A',y='B',kind='scatter')


plt.title('Linear Approximation of B function')
plt.xlabel('A')
plt.ylabel('B')
plt.legend(loc='upper left')
plt.grid()

m, c = np.polyfit(df['A'],df['B'], 1)
plt.plot(df['A'],m*df['A']+c,'r',label='Function B Approximation(x) = m*x + b')
#plt.plot(df['A'],m*np.exp(df['A']),'r',label='Function B Approximation(x) = m*e^A + c')
plt.show()


```
