```python
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx
import random
from collections import Counter
from nltk.corpus import stopwords
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
```

```python
df = pd.read_csv('philly_users_businesses_stars.csv')

chunks = pd.read_json('yelp_academic_dataset_review.json', lines=True,chunksize=100000)
df_reviews = pd.DataFrame()
dataProcessed = 0
for chunk in chunks:
    dataProcessed+=1
    #temp = pd.merge(left=df, right=chunk, on='user_id')
    temp = chunk.loc[chunk['user_id'].isin(df['user_id'])]
    print(f'data processed = {dataProcessed*100000}')
    df_reviews = pd.concat([df_reviews,temp])

```

```python
print(df)
print(df_reviews.columns)
df_reviews = df_reviews[['stars','text']]


good_reviews = df_reviews.loc[df_reviews['stars']>=5].head(1000)
bad_reviews = df_reviews.loc[df_reviews['stars']<=1].head(1000)


good_merged_text = ''.join(good_reviews['text'])
bad_merged_text = ''.join(bad_reviews['text'])

good_words_tokenized = word_tokenize(good_merged_text)
bad_words_tokenized = word_tokenize(bad_merged_text)

good_filtered_words = [word for word in good_words_tokenized if word.lower() not in stopwords.words('english')]
bad_filtered_words = [word for word in bad_words_tokenized if word.lower() not in stopwords.words('english')]


good_words_without_stopwords = ' '.join(good_filtered_words)
bad_words_without_stopwords = ' '.join(bad_filtered_words)



good_words = good_words_without_stopwords.split()
bad_words = bad_words_without_stopwords.split()

good_word_counts = Counter(good_words)
bad_word_counts = Counter(bad_words)

#apo ta stopwords den aferaithikan polla simeia stiksis kai alles lekseis opws " 've " 
# paratiroume oti apo tis 'kanonikes' lekseis arketes einai endeiktikes oti ena estiatorio
# arese se enan user opws great,good,like,best klp

#gia to bad leksiko itan pio dyskolo mias kai polles fores emfanistike px i leksi "like"
#omws kata pasa pithanotita proigoutan ena didn't
 
print(good_word_counts)
print(bad_word_counts)

```
